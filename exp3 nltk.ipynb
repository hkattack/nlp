```python
import requests
from nltk import FreqDist
from nltk.corpus import stopwords
import seaborn as sns
%matplotlib inline
```

```python
url="https://www.gutenberg.org/files/11/11-0.txt"
alice=requests.get(url)
print(alice.text)
```

```python
def plot_word_frequency(words,top_n=10):
    word_freq=FreqDist(words)
    labels=[element[0]for element in word_freq.most_common(top_n)]
    counts=[element[1]for element in word_freq.most_common(top_n)]
    sns.set(rc={'figure.figsize':(15,8)})
    plot=sns.barplot(x=labels,y=counts)
    return plot
```

```python
alice_words=alice.text.split()
plot_word_frequency(alice.text,15)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
example_sent="My name is Kunal Chawla. I am a student at Chandigarh University."
stop_words=set(stopwords.words('english'))
filtered_sentence=[a for a in alice_words if not a in stop_words]
filtered_sentence=[]
for a in alice_words:
    if a not in stop_words:
        filtered_sentence.append(a)
plot_word_frequency(filtered_sentence,15)
```

Stopwords are common words in a language that are often considered insignificant or do not carry much meaning in the context of text analysis. These words, such as "the," "is," "and," "in," etc., occur frequently and can be found in almost every document. In many NLP tasks, stopwords are removed to improve processing efficiency and focus on more meaningful words.

To reduce stopwords in natural language, you can follow these steps:

1. Stopword List: Obtain a list of stopwords specific to your language or domain. Many NLP libraries, such as NLTK for Python, provide built-in stopwords lists for various languages.

2. Tokenization: Tokenize the text into individual words or tokens.

3. Filtering: Compare each word against the stopwords list and remove any word that matches a stopword.

4. Contextual Analysis: Consider the context and the specific NLP task at hand. Some stopwords may carry important meaning depending on the analysis. For example, in sentiment analysis, words like "not" or "no" may be important for understanding negative sentiments, so they may not be removed as stopwords.

5. Customization: If the provided list of stopwords does not suit your needs, you can create a custom stopwords list by adding or removing words according to your requirements.

It's important to note that the removal of stopwords may not be suitable or necessary for all NLP tasks. In some cases, stopwords can provide valuable information, such as in language modeling or topic modeling, where the frequency of words is important.

By reducing or removing stopwords, you can often improve the efficiency of downstream NLP tasks, such as text classification, sentiment analysis, information retrieval, and more, by focusing on the more informative and meaningful words.
